## クラスタアルゴリズム（10）

```python
from numpy import unique
from numpy import where

from sklearn.datasets import make_classification
```

```python
# define dataset
X, _ = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)
X
```

```python
def plt_cluster(cluster):
    for cluster in clusters:
        row_ix = where(yhat == cluster)
        plt.scatter(X[row_ix, 0], X[row_ix, 1])
    plt.show()
```

```python
cluster_num = 5
```

#### 1. Affinity Propagation
アフィニティ伝播には、データを最適に要約する一連の見本を見つけることが含まれます。

```python
from sklearn.cluster import AffinityPropagation
```

```python
# define the model
model = AffinityPropagation(damping=0.9)

# fit the model
model.fit(X)

# assign a cluster to each example
yhat = model.predict(X)

# retrieve unique clusters
clusters = unique(yhat)

# show plot
plt_cluster(clusters)
```

#### 2. Agglomerative Clustering
凝集型クラスタリングでは、必要な数のクラスターが達成されるまで例をマージします。

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# define the model
model = AgglomerativeClustering(n_clusters=cluster_num)

# fit model and predict clusters
yhat = model.fit_predict(X)

# retrieve unique clusters
clusters = unique(yhat)

# show plot
plt_cluster(clusters)
```

#### 3. BIRCH
BIRCHクラスタリング（BIRCHは、階層を使用したBalanced Iterative Reducing and Clusteringの略です）には、クラスターの重心が抽出されるツリー構造の構築が含まれます。

```python
from sklearn.cluster import Birch
```

```python
# define the model
model = Birch(threshold=0.01, n_clusters=cluster_num)

# fit the model
model.fit(X)

# assign a cluster to each example
yhat = model.predict(X)

# retrieve unique clusters
clusters = unique(yhat)

# show plot
plt_cluster(clusters)
```

#### 4. DBSCAN
DBSCANクラスタリングは、ドメイン内の高密度領域を見つけ、それらの周りの特徴空間のそれらの領域をクラスターとして拡張する

```python
from sklearn.cluster import DBSCAN
```

```python
# define the model
model = DBSCAN(eps=0.30, min_samples=9)

# fit model and predict clusters
yhat = model.fit_predict(X)

# retrieve unique clusters
clusters = unique(yhat)

# show plot
plt_cluster(clusters)
```

#### 5. K-Means
K平均法は、最も広く知られているクラスタリングアルゴリズムであり、各クラスター内の分散を最小限に抑えるクラスターを生成する

```python
from sklearn.cluster import KMeans
```

```python
# define the model
model = KMeans(n_clusters=cluster_num)

# fit the model
model.fit(X)

# assign a cluster to each example
yhat = model.predict(X)

# retrieve unique clusters
clusters = unique(yhat)

# show plot
plt_cluster(clusters)
```

#### 6. Mini-Batch K-Means
ミニバッチK平均法は、データセット全体ではなくサンプルのミニバッチを使用してクラスターの重心を更新するk平均法の修正バージョンです。これにより、大規模なデータセットでより高速になり、統計ノイズに対してより堅牢になる可能性があります。

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
# define the model
model = MiniBatchKMeans(n_clusters=cluster_num)

# fit the model
model.fit(X)

# assign a cluster to each example
yhat = model.predict(X)

# retrieve unique clusters
clusters = unique(yhat)

# show plot
plt_cluster(clusters)
```

#### 7. Mean Shift
平均シフトクラスタリングには、特徴空間内の例の密度に基づいて重心を見つけて適応させることが含まれます。

```python
from sklearn.cluster import MeanShift
```

```python
# define the model
model = MeanShift()

# fit model and predict clusters
yhat = model.fit_predict(X)

# retrieve unique clusters
clusters = unique(yhat)

# show plot
plt_cluster(clusters)
```

#### 8. OPTICS
OPTICSクラスタリング（OPTICSは、クラスタリング構造を特定するための順序付けポイントの略）は、上記のDBSCANの修正バージョンです。

```python
from sklearn.cluster import OPTICS
```

```python
# define the model
model = OPTICS(eps=0.8, min_samples=10)

# fit model and predict clusters
yhat = model.fit_predict(X)

# retrieve unique clusters
clusters = unique(yhat)

# show plot
plt_cluster(clusters)
```

#### 9. Spectral Clustering
スペクトルクラスタリングは、線形代数から引き出されたクラスタリング手法の一般的なクラスです。

```python
from sklearn.cluster import SpectralClustering
```

```python
# define the model
model = SpectralClustering(n_clusters=cluster_num)

# fit model and predict clusters
yhat = model.fit_predict(X)

# retrieve unique clusters
clusters = unique(yhat)

# show plot
plt_cluster(clusters)
```

#### 10. Gaussian Mixture Model
ガウス混合モデルは、その名前が示すように、ガウス確率分布の混合を伴う多変量確率密度関数を要約します。

```python
from sklearn.mixture import GaussianMixture
```

```python
# define the model
model = GaussianMixture(n_components=cluster_num)

# fit the model
model.fit(X)

# assign a cluster to each example
yhat = model.predict(X)

# retrieve unique clusters
clusters = unique(yhat)

# show plot
plt_cluster(clusters)
```
